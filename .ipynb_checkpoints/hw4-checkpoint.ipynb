### 1. [20 pts] RNN basics

- In your own words, explain the difference between CNNs and RNNs comprehensively.
- How do RNNs store past information?
- In RNNs, why do we use the same 𝑊at every time step instead of different matrices 𝑊_1,𝑊_2,...?
- What information do we need to compute the hidden state at time t?


Answer:

- CNNs are mainly used for spatial data like images. They look for patterns using filters over local regions and don’t really care about order. RNNs are designed for sequential data like text or time series, where order matters. They process input step-by-step and keep track of previous information through a hidden state.
- RNNs store past info in the hidden state, which gets updated at every time step and passed forward, so it kind of “remembers” previous inputs.
- We use the same weights to keep the model consistent across time and reduce the number of parameters. If we used different weights each step, it would be too complex and wouldn’t generalize well.
- We need the current input, the previous hidden state and the weight matrices (plus bias)

### 2. [20 pts] LSTM basics

- Why LSTMs have four gates?
- What's the cell state in a LSTM model?
- What is gradient vanishing? What is gradient exploding?
- Do LSTMs solve the vanishing gradient problem?

Answer:

- They have gates to control what information to keep, forget, and output. Specifically: forget gate, input gate, candidate values, and output gate. This helps manage long-term dependencies better.
- The cell state is like a memory that runs through the entire sequence. It carries important information forward with minimal changes.
- Vanishing gradient means gradients get really small during backprop, so the model stops learning long-term dependencies. Exploding gradient is when gradients get too large and cause unstable updates.
- They don’t completely solve it, but they significantly reduce the problem using gates and the cell state.

### 3. [20 pts] Representation Learning

- In your own words, explain the difference between pre-training and fine-tuning.
- What is transfer learning?


Answer:

- Pre-training is when a model is trained on a large general dataset to learn basic patterns. Fine-tuning is when you take that model and train it more on a specific task or dataset.
- Transfer learning is using knowledge learned from one task (usually from pre-training) and applying it to another related task to improve performance and reduce training time.

### 4. [20 pts] Transformer basics: compute attention in code

Write code to compute the self-attention output for token 1 in a single-head self-attention layer.

Given:
- $Q_1 = [1, 0]$
- $K_1 = [1, 0]$, $V_1 = [2, 0]$
- $K_2 = [0, 1]$, $V_2 = [0, 2]$
- $K_3 = [1, 1]$, $V_3 = [1, 1]$

Use scaled dot-product attention with $d_k = 2$:
$$
\mathrm{Attention}(Q, K, V) = \mathrm{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V
$$

Your code should:
- compute the raw attention scores from $Q_1$ to each key
- scale the scores by $\sqrt{2}$
- apply softmax
- compute the final weighted sum of the value vectors

Print:
- the raw scores
- the scaled scores
- the attention weights
- the final attention output

You should report approximate values rounded to 3 decimals.


```python
import math

Q1 = [1.0, 0.0]
K = [
    [1.0, 0.0],
    [0.0, 1.0],
    [1.0, 1.0],
]
V = [
    [2.0, 0.0],
    [0.0, 2.0],
    [1.0, 1.0],
]

d_k = len(K[0])

# TODO 1: compute the raw dot-product scores from Q1 to each key in K
raw_scores = []
for key in K:
    score = Q1[0] * key[0] + Q1[1] * key[1]
    raw_scores.append(score)
    
# TODO 2: scale the scores by sqrt(d_k)
scaled_scores = []
for score in raw_scores:
    scaled_scores.append(score / math.sqrt(d_k))
    
# TODO 3: apply softmax to get attention weights
exp_scores = [math.exp(score) for score in scaled_scores]
sum_exp = sum(exp_scores)
attention_weights = [score / sum_exp for score in exp_scores]

# TODO 4: compute the weighted sum of the value vectors
attention_output = [0.0, 0.0]
for weight, value in zip(attention_weights, V):
    attention_output[0] += weight * value[0]
    attention_output[1] += weight * value[1]

raw_scores = [round(x, 3) for x in raw_scores]
scaled_scores = [round(x, 3) for x in scaled_scores]
attention_weights = [round(x, 3) for x in attention_weights]
attention_output = [round(x, 3) for x in attention_output]

print('Raw scores:', raw_scores)
print('Scaled scores:', scaled_scores)
print('Attention weights:', attention_weights)
print('Attention output:', attention_output)
```

### 5. [20 pts] Reinforcement learning basics: compute Q-values in code

Consider the following deterministic environment arranged as a $2 \times 2$ grid:

| $S_0$ | $S_1$ |
|---|---|
| $S_2$ | Goal |

Allowed transitions:
- from $S_0$: `RIGHT -> S_1`, `DOWN -> S_2`
- from $S_1$: `LEFT -> S_0`, `DOWN -> Goal`
- from $S_2$: `UP -> S_0`, `RIGHT -> Goal`
- `Goal` is terminal

Use the Bellman optimality equation:
$$
Q(s, a) = r + \gamma \max_{a'} Q(s', a')
$$

Assume:
- discount factor $\gamma = 1$
- every action gives reward $-1$
- entering `Goal` also gives reward $-1$
- the terminal state's future value is $0$

Using Python, write code to compute the following values:
- $Q(S_1, \mathrm{DOWN})$
- $Q(S_2, \mathrm{RIGHT})$
- $Q(S_0, \mathrm{RIGHT})$
- $Q(S_0, \mathrm{DOWN})$

Your code should:
- represent the environment transitions
- compute the optimal state values using value iteration
- compute the requested Q-values from those optimal values
- print the final state values and the four requested Q-values

You should report exact integer values.


```python
states = ['S0', 'S1', 'S2', 'Goal']
actions = {
    'S0': {'RIGHT': 'S1', 'DOWN': 'S2'},
    'S1': {'LEFT': 'S0', 'DOWN': 'Goal'},
    'S2': {'UP': 'S0', 'RIGHT': 'Goal'},
    'Goal': {},
}

gamma = 1
reward = -1

# Start with all state values at 0.
V = {state: 0 for state in states}

for iteration in range(10):
    new_V = V.copy()
    for state in states:
        if state == 'Goal':
            new_V[state] = 0
            continue

        q_values = []
        for next_state in actions[state].values():
            # TODO 1: implement the core Bellman update
            q = reward + gamma * V[next_state]
            q_values.append(q)

        new_V[state] = max(q_values)

    V = new_V

def compute_q(state, action):
    # TODO 2: complete the final q value computation function
    next_state = actions[state][action]
    return reward + gamma * V[next_state]

print('Final state values:', V)

print('Q(S1, DOWN) =', compute_q('S1', 'DOWN'))
print('Q(S2, RIGHT) =', compute_q('S2', 'RIGHT'))
print('Q(S0, RIGHT) =', compute_q('S0', 'RIGHT'))
print('Q(S0, DOWN) =', compute_q('S0', 'DOWN'))
```
